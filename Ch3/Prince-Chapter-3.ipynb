### Problem 2
To find the peak (mode) of the probability distribution, we need to find the maximum of the probability distribution function:

\begin{aligned}
Pr(\lambda) = \frac{\Gamma[\alpha + \beta]}{\Gamma[\alpha] \Gamma[\beta]} \lambda^{\alpha - 1}(1-\lambda)^{\beta-1}
\end{aligned}

For the $\alpha, \beta$ values given, the pdf has a single extremum point, which corresponds to the mode of the distribution.  To find the value of the mode, differentiate $Pr(\lambda)$ w.r.t. $\lambda$ and then evaluate where the derivative is equal to zero.

\begin{aligned}
\frac{d Pr(\lambda)}{d \lambda} &= \frac{\Gamma[\alpha + \beta]}{\Gamma[\alpha] \Gamma[\beta]} \bigl ( (\alpha - 1) \lambda^{\alpha - 2}(1-\lambda)^{\beta-1} + (\beta-1)\lambda^{\alpha - 1}(1-\lambda)^{\beta-2} \bigr ) \\
&= 0
\end{aligned}

The derivative equal to 0 implies

\begin{aligned}
0 &= (\alpha - 1) \lambda^{\alpha - 2}(1-\lambda)^{\beta-1} + (\beta-1)\lambda^{\alpha - 1}(1-\lambda)^{\beta-2} \\
&= \lambda^{\alpha-2} (1-\lambda)^{\beta-2} \bigl ( (\alpha - 1) (1 - \lambda) + (\beta-1) \lambda \bigr )
\end{aligned}

$\lambda$ is non-zero, so we can divide it out of the rhs:

\begin{aligned}
0 &= (\alpha - 1) (1 - \lambda) + (\beta-1) \lambda \\
&= \alpha (1 - \lambda) - 1 (1 - \lambda) + (\beta-1) \lambda \\
&= (\alpha - 1) + (\beta - \alpha) \lambda
\end{aligned}

which yields $\lambda = \frac{\alpha - 1}{\alpha - \beta}$.

### Problem 3
Given:

\begin{aligned}
\mu &= \frac{\alpha}{\alpha+\beta} \\
\sigma^2 &= \frac{\alpha \beta}{(\alpha+\beta)^2(\alpha+\beta+1)}
\end{aligned}

we observe that $\mu(1-\mu) = \frac{\alpha \beta}{(\alpha+\beta)^2}$.  Combined with the given relationships, we have:

\begin{aligned}
\alpha + \beta + 1 &= \frac{\mu(1-\mu)}{\sigma^2} \\
\implies \alpha + \beta &= \frac{\mu(1-\mu)}{\sigma^2} - 1
\end{aligned}

Now, we can solve for $\alpha$:

\begin{aligned}
\mu &= \frac{\alpha}{\alpha+\beta} \\
\implies \alpha &= \mu(\alpha + \beta) \\
&= \mu \biggl ( \frac{\mu(1-\mu)}{\sigma^2} - 1 \biggr )
\end{aligned}

For $\beta$:

\begin{aligned}
\beta &= \frac{\mu(1-\mu)}{\sigma^2} - \alpha - 1 \\
&= \frac{\mu(1-\mu)}{\sigma^2} -1 - \mu \biggl ( \frac{\mu(1-\mu)}{\sigma^2} - 1 \biggr ) \\
&= \bigl ( 1 - \mu \bigr ) \biggl ( \frac{\mu(1-\mu)}{\sigma^2} - 1 \biggr )
\end{aligned}

### Problem 5
Let $u = t^z$ and $dv = e^{-t}dt$.  Then $\Gamma[z+1]$ can be evaluated as:

\begin{aligned}
\Gamma[z+1] = \int_0^\infty u dv &= uv \bigl |_0^\infty - \int_0^\infty vdu \\
&\triangleq -t^z e^{-t} \bigl |_0^\infty + \int_0^\infty e^{-t} z t^{z-1} dt \\

&= 0 \cdot 1 - \lim_{t \to \infty} t^z e^{-t} + z \Gamma[z] \\
&= 0 \cdot 1 - 0 + z \Gamma[z] \\
&= z \Gamma[z]
\end{aligned}

### Problem 8
The mode occurs when:
\begin{aligned}
\frac{\partial Pr(\mu, \sigma^2)}{\partial \mu} &= 0 \\
\frac{\partial Pr(\mu, \sigma^2)}{\partial \sigma^2} &= 0
\end{aligned}
The probability density function is:
\begin{aligned}
Pr(\mu, \sigma^2) &= \frac{\sqrt{\gamma}}{\sqrt{2 \pi \sigma^2}} \frac{\beta^\alpha}{\Gamma[\alpha]} \Bigl ( \frac{1}{\sigma^2} \Bigr )^{\alpha + 1} \exp \biggl [ -\frac{2 \beta + \gamma(\delta-\mu)^2}{2 \sigma^2} \biggr ]
\end{aligned}
The derivatives are most easily evaluated with `sympy` (see code cell below).  The derivatives are equal to zero when the following is true:
\begin{aligned}
\delta - \mu = 0 &\implies& \frac{\partial Pr(\mu, \sigma^2)}{\partial \mu} &= 0 \\
2 \beta + \gamma(\delta - \mu)^2 - \Bigl ( \frac{1}{\sigma^2} \Bigr )^{\alpha - 3/2} (2 \alpha + 3) = 0 &\implies& \frac{\partial Pr(\mu, \sigma^2)}{\partial \sigma^2} &= 0
\end{aligned}
$\delta = \mu$ from the first relation forces a term in the second to drop out:
\begin{aligned}
2 \beta - \Bigl ( \frac{1}{\sigma^2} \Bigr )^{\alpha - 3/2} (2 \alpha + 3) &= 0 \\
\implies \sigma^2 &= \Bigl ( \frac{2 \beta}{2 \alpha + 3} \Bigr )^{\frac{1}{3/2 - \alpha}}
\end{aligned}

```python
from sympy import *
# x == mu, y == sigma**2
x, y = symbols('x y')
# a == alpha, b == beta, c == (sqrt(gamma) * beta**alpha) / (sqrt(2*pi) * \Gamma(alpha)), d == delta, g == gamma 
a, b, c, d, g = symbols('a b c d g')
Pr = c / sqrt(y) * y**(-(a+1)) * exp(-(2*b + g*(d-x)**2)/(2*y))
print(simplify(diff(Pr, x)))
print(simplify(diff(Pr, y)))
```
