### Problem 1

\begin{aligned}
\mathbf{y} &= A \mathbf{x} + \mathbf{b} \\
\mathbf{x} &= A^{-1} (\mathbf{y} - \mathbf{b})
\end{aligned}

By the change-of-variables formula:

\begin{aligned}
Pr(\mathbf{x}) &= |A^{-1}| Pr(A^{-1} (\mathbf{y} - \mathbf{b})) \\
&= \frac{1}{(2 \pi)^{D/2} |\Sigma| |A|} \exp \bigl[ -0.5 (A^{-1} (\mathbf{y} - \mathbf{b}) - \mathbf{\mu} )^T \Sigma^{-1} (A^{-1} (\mathbf{y} - \mathbf{b}) - \mathbf{\mu} ) \bigr ]
\end{aligned}

Start with the argument to the exponential function:

\begin{aligned}
A^{-1}(\mathbf{y} - \mathbf{b}) - \mu &= A^{-1}(\mathbf{y} - \mathbf{b} - A \mu) \\
&= A^{-1}(\mathbf{y} - (A \mu + \mathbf{b})) \\
(A^{-1}(\mathbf{y} - (A \mu + \mathbf{b})))^T \Sigma^{-1} (A^{-1}(\mathbf{y} - (A \mu + \mathbf{b}))) &= (\mathbf{y} - (A \mu + \mathbf{b}))^T A^{-T} \Sigma^{-1} A^{-1} (\mathbf{y} - (A \mu + \mathbf{b})) \\
&= (\mathbf{y} - (A \mu + \mathbf{b}))^T (A \Sigma A^T)^{-1} (\mathbf{y} - (A \mu + \mathbf{b})).
\end{aligned}

Now, see if the normalizer can be manipulated to get the desired $|A \Sigma A^T|^{1/2}$ term in the denominator:

\begin{aligned}
\frac{1}{(2 \pi)^{D/2} |\Sigma| |A|} &= \frac{1}{(2 \pi)^{D/2} |\Sigma| |A|^{1/2} |A|^{1/2} } \\
&= \frac{1}{(2 \pi)^{D/2} |\Sigma| |A|^{1/2} |A^T|^{1/2} } \\
&= \frac{1}{(2 \pi)^{D/2} |A|^{1/2} |\Sigma| |A^T|^{1/2}} \\
&= \frac{1}{(2 \pi)^{D/2} |A \Sigma A^T|^{1/2}}
\end{aligned}

Putting both parts together:

\begin{aligned}
Pr(\mathbf{y}) &= \frac{1}{(2 \pi)^{D/2} |A \Sigma A^T|^{1/2}} \exp \bigl [ -0.5 (\mathbf{y} - (A \mu + \mathbf{b}))^T (A \Sigma A^T)^{-1} (\mathbf{y} - (A \mu + \mathbf{b})) \bigr ]
\end{aligned}


### Problem 2
Apply results of Problem 1 with $A = \Sigma^{-1/2}$ and $\mathbf{b} = -\Sigma^{-1/2} \mathbf{\mu}$.

\begin{aligned}
\mathbf{\bar \mu} &= A \mathbf{\mu} + \mathbf{b} \\
&= \Sigma^{-1/2} \mathbf{\mu} -\Sigma^{-1/2} \mathbf{\mu} \\
&= \mathbf{0} \\
\bar \Sigma &= A \Sigma A^T \\
&= \Sigma^{-1/2} \Sigma (\Sigma^{-1/2})^T \\
\text{(Cholesky decomposition)} &= \Sigma^{-1/2} \Sigma^{1/2} (\Sigma^{1/2})^T (\Sigma^{-1/2})^T \\
&= \Sigma^{-1/2} \Sigma^{1/2} (\Sigma^{-1/2} \Sigma^{1/2})^T \\
&= I \cdot I \\
&= I
\end{aligned}

### Problem 4
Let:

\begin{aligned}
\begin{bmatrix}
U & V \\
W & X
\end{bmatrix}
\equiv \begin{bmatrix}
(A - BD^{-1}C)^{-1} & -(A-BD^{-1}C)^{-1} BD^{-1} \\
-D^{-1}C(A - BD^{-1}C)^{-1} & D^{-1} + D^{-1}C(A - BD^{-1}C)^{-1} BD^{-1} 
\end{bmatrix}
\end{aligned}

Now, show that:

\begin{aligned}
\begin{bmatrix}
U & V \\
W & X
\end{bmatrix} \begin{bmatrix}
A & B \\
C & D
\end{bmatrix} &= \begin{bmatrix}
I & 0 \\
0 & I
\end{bmatrix}
\end{aligned}

Do the calculation block-by-block.

##### Block (1, 1): $UA + VC = I$

\begin{aligned}
(A - BD^{-1}C)^{-1} A -(A-BD^{-1}C)^{-1} BD^{-1} C &= (A - BD^{-1}C)^{-1} (A - BD^{-1} C) \\
&= I
\end{aligned}

##### Block (1, 2): $UB + VD = 0$

\begin{aligned}
(A - BD^{-1}C)^{-1} B - (A-BD^{-1}C)^{-1} BD^{-1} D &= (A - BD^{-1}C)^{-1} B - (A-BD^{-1}C)^{-1} B \\
&= 0
\end{aligned}

##### Block (2, 1): $WA + XC = 0$

\begin{aligned}
-D^{-1}C(A - BD^{-1}C)^{-1} A + (D^{-1} + D^{-1}C(A - BD^{-1}C)^{-1} BD^{-1})C &= D^{-1} (-C(A - BD^{-1}C)^{-1} A + I + C(A - BD^{-1}C)^{-1} BD^{-1})C \\
&= D^{-1} [ -C (A - BD^{-1}C)^{-1} (A - BD^{-1}C)) + CI ] \\
&= D^{-1} C (-I + I) \\
&= 0
\end{aligned}

##### Block(2, 2): $WB + XD = I$:

\begin{aligned}
-D^{-1}C(A - BD^{-1}C)^{-1} B + (D^{-1} + D^{-1}C(A - BD^{-1}C)^{-1} BD^{-1})D &= D^{-1} (-C(A - BD^{-1}C)^{-1} B + I + C(A - BD^{-1}C)^{-1} BD^{-1})D \\
&= D^{-1} [- C(A - BD^{-1}C)^{-1} B + D + C(A - BD^{-1}C)^{-1} B)]\\
&= D^{-1} D \\
&= I
\end{aligned}

Use the result of Problem 4, with:

-\Sigma_{22}^{-1} \Sigma_{21} (\Sigma_{11} - \Sigma_{21}^T \Sigma_{22}^{-1}\Sigma_{21})^{-1} & \Sigma_{22}^{-1} + \Sigma_{22}^{-1} \Sigma_{21} (\Sigma_{11} - \Sigma_{21}^T \Sigma_{22}^{-1}\Sigma_{21})^{-1} \Sigma_{21}^T \Sigma_{22}^{-1}

&\triangleq
\begin{bmatrix}
U & V \\
W & X
\end{bmatrix}
\end{aligned}

$U$ appears in the off-diagonal block terms:

\begin{aligned}
V &= -U \Sigma_{21}^T \Sigma_{22}^{-1} \\
W &= V^T

Pr( \mathbf{x} ) &= \kappa_1 \exp \biggl [ -0.5 \begin{bmatrix} \mathbf{x}_1 - \mathbf{\mu}_1 \\ \mathbf{x}_2 - \mathbf{\mu}_2 \end{bmatrix}^T \begin{bmatrix} U & W^T \\ W & X \end{bmatrix} \begin{bmatrix} \mathbf{x}_1 - \mathbf{\mu}_1 \\ \mathbf{x}_2 - \mathbf{\mu}_2 \end{bmatrix} \biggr ] \\
&= \kappa_1 \exp \Bigl [ -0.5 \bigl [(\mathbf{x}_1 - \mathbf{\mu}_1)^T U (\mathbf{x}_1 - \mathbf{\mu}_1) + (\mathbf{x}_2 - \mathbf{\mu}_2)^T W (\mathbf{x}_1 - \mathbf{\mu}_1) \\
&+& (\mathbf{x}_1 - \mathbf{\mu}_1)^T W^T (\mathbf{x}_2 - \mathbf{\mu}_2) + (\mathbf{x}_2 - \mathbf{\mu}_2)^T X (\mathbf{x}_2 - \mathbf{\mu}_2) \bigr ] \Bigr ] \\

\text{cts} &\triangleq \kappa_3 \exp \Bigl [ -0.5 \bigl [(\mathbf{x}_1 - \mathbf{\mu}_1)^T U (\mathbf{x}_1 - \mathbf{\mu}_1) + (\mathbf{x}_2 - \mathbf{\mu}_2)^T W (\mathbf{x}_1 - \mathbf{\mu}_1) + (\mathbf{x}_1 - \mathbf{\mu}_1)^T W^T (\mathbf{x}_2 - \mathbf{\mu}_2) + (\mathbf{x}_2 - \mathbf{\mu}_2)^T W U^{-1} W^T (\mathbf{x}_2 - \mathbf{\mu}_2)\bigr ] \Bigr ] \\
&= \kappa_3 \exp \Bigl [ -0.5 \bigl [\bigl (\mathbf{x}_1 - (\mathbf{\mu_1} - U^{-1}W^T [\mathbf{x}_2 - \mathbf{\mu_2}]) \bigr )^T U \bigl (\mathbf{x}_1 - (\mathbf{\mu_1} - U^{-1}W^T [\mathbf{x}_2 - \mathbf{\mu_2}]) \bigr ) \bigr ) \bigr ] \Bigr ]
\end{aligned}

But, the joint density $Pr(\mathbf{x}_1, \mathbf{x}_2) \equiv Pr(\mathbf{x})$.  It follows, given the fact that $Pr(\mathbf{x}) Pr(\mathbf{x}_2) = Pr(\mathbf{x}_1 | \mathbf{x}_2)$, that

\begin{aligned}
Pr(\mathbf{x}_1 | \mathbf{x}_2) &= \frac{Pr(\mathbf{x})}{Pr(\mathbf{x}_2)} \\
&= \frac{\kappa_3}{Pr(\mathbf{x}_2)} \exp \Bigl [ -0.5 \bigl [\bigl (\mathbf{x}_1 - (\mathbf{\mu_1} - U^{-1}W^T [\mathbf{x}_2 - \mathbf{\mu_2}]) \bigr )^T U \bigl (\mathbf{x}_1 - (\mathbf{\mu_1} - U^{-1}W^T [\mathbf{x}_2 - \mathbf{\mu_2}]) \bigr ) \bigr ) \bigr ] \Bigr ]
\end{aligned}

$Pr(\mathbf{x}_1 | \mathbf{x}_2)$ is a valid probability distribution, so it must be true that

\begin{aligned}
\frac{\kappa_3}{Pr(\mathbf{x}_2)} &= \frac{1}{(2 \pi)^{D/2} |U^{-1}|^{1/2}}
\end{aligned}

Finally, we can conclude (substituting $U, W$ from above):

\begin{aligned}
Pr(\mathbf{x}_1 | \mathbf{x}_2) &= \text{Norm}_{\mathbf{x}_1} \Bigl [ \mathbf{\mu_1} + (\Sigma_{11} - \Sigma_{21}^T \Sigma_{22}^{-1}\Sigma_{21}) (\Sigma_{11} - \Sigma_{21}^T \Sigma_{22}^{-1}\Sigma_{21})^{-1} \Sigma_{21}^T \Sigma_{22}^{-1} (\mathbf{x}_2 - \mathbf{\mu_2}), \Sigma_{11} - \Sigma_{21}^T \Sigma_{22}^{-1}\Sigma_{21} \Bigr ] \\
&= \text{Norm}_{\mathbf{x}_1} \Bigl [ \mathbf{\mu_1} + \Sigma_{21}^T \Sigma_{22}^{-1} (\mathbf{x}_2 - \mathbf{\mu_2}), \Sigma_{11} - \Sigma_{21}^T \Sigma_{22}^{-1}\Sigma_{21} \Bigr ] \\
\end{aligned}

### Problem 7
Start with the product:

\begin{aligned}
\text{Norm}_{\mathbf{x}} [\mathbf{a}, A ] \cdot \text{Norm}_{\mathbf{x}} [\mathbf{b}, B ] 
\end{aligned}

Combine all normalization constants into $\kappa_1$:

\begin{aligned}
\text{Norm}_{\mathbf{x}} [\mathbf{a}, A ] \cdot \text{Norm}_{\mathbf{x}} [\mathbf{b}, B ] &= \kappa_1 \exp \bigl [ -0.5 (\mathbf{x} - \mathbf{a})^T A^{-1} (\mathbf{x} - \mathbf{a}) \bigr ] \cdot \exp \bigl [ -0.5 (\mathbf{x} - \mathbf{b})^T B^{-1} (\mathbf{x} - \mathbf{b}) \bigr ] \\

\end{aligned}

Note that, since $A$ is a covariance matrix, $A^T = A, A^{-T} = A^{-1}$ and it follows that:

### Problem 8

\begin{aligned}
\text{Norm}_x [\mu_1, \sigma_1^2 ] \cdot \text{Norm}_x [\mu_2, \sigma_2^2] &= \frac{1}{2 \pi \sqrt{\sigma_1^2 \sigma_2^2}} \exp [-\frac{1}{2 \sigma_1^2} (x - \mu_1)^2 - \frac{1}{2 \sigma_2^2} (x - \mu_2)^2] \\
&= \frac{1}{2 \pi \sqrt{\sigma_1^2 \sigma_2^2}} \exp \Bigl [ -\frac{1}{2 \sigma_1^2 \sigma_2^2} \bigl [ \sigma_2^2 (x - \mu_1)^2 - \sigma_1^2 (x - \mu_2)^2 \bigr ] \Bigr ] \\
&= \frac{1}{2 \pi \sqrt{\sigma_1^2 \sigma_2^2}} \exp \Bigl [ -\frac{1}{2 \sigma_1^2 \sigma_2^2} \bigl [ (\sigma_1^2 + \sigma_2^2) x^2 - 2 (\mu_1 \sigma_2^2 + \mu_2 \sigma_1^2)x + \mu_1^2 \sigma_1^2 + \mu_2 \sigma_2^2 \bigr ] \Bigr ] \\
&\triangleq \kappa_1 \exp \Bigl [ -\frac{\sigma_1^2 + \sigma_2^2}{2 \sigma_1^2 \sigma_2^2} \bigl [ x^2 - 2 \frac{\mu_1 \sigma_2^2 + \mu_2 \sigma_1^2}{\sigma_1^2 + \sigma_2^2} x \bigr ] \Bigr ] \\
&\triangleq \kappa_2 \exp \Bigl [ -\frac{\sigma_1^2 + \sigma_2^2}{2 \sigma_1^2 \sigma_2^2} \bigl [ x^2 - 2 \frac{\mu_1 \sigma_2^2 + \mu_2 \sigma_1^2}{\sigma_1^2 + \sigma_2^2} x + \bigl ( \frac{\mu_1 \sigma_2^2 + \mu_2 \sigma_1^2}{\sigma_1^2 + \sigma_2^2} \bigr )^2 \bigr ] \Bigr ] \\
&= \kappa_2 \exp \Bigl [ -\frac{\sigma_1^2 + \sigma_2^2}{2 \sigma_1^2 \sigma_2^2} \bigl [ \bigl ( x - \frac{\mu_1 \sigma_2^2 + \mu_2 \sigma_1^2}{\sigma_1^2 + \sigma_2^2} \bigr )^2 \bigr ] \Bigr ] \\
&\propto& \text{Norm}_x \Bigl [ \frac{\mu_1 \sigma_2^2 + \mu_2 \sigma_1^2}{\sigma_1^2 + \sigma_2^2}, \frac{\sigma_1^2 \sigma_2^2}{\sigma_1^2 + \sigma_2^2} \Bigr ]
\end{aligned}

#### Show that $\mu_1 \le \mu \le \mu_2$:

Assume, WLOG, that $\mu_1 \le \mu_2$.  It remains to show that $\mu_1 \le \mu \le \mu_2$, where:

\begin{aligned}
\mu &\triangleq \frac{\mu_1 \sigma_2^2 + \mu_2 \sigma_1^2}{\sigma_1^2 + \sigma_2^2} \\
&= a \mu_1 + (1-a) \mu_2 \\
a &= \frac{\sigma_2^2}{\sigma_1^2 + \sigma_2^2} \le 1
\end{aligned}

Clearly, $\mu \le \mu_2$, since

\begin{aligned}
\mu &= \mu_2 + a (\mu_1 - \mu_2) \\
&\le& \mu_2
\end{aligned}

Since $\mu_1 \le \mu_2$ by assumption, it follows that

\begin{aligned}
(1-a) \mu_2 &\ge& (1-a) \mu_1 \\
(1-a) \mu_2 &\ge& (1-a) \mu_1 \\
\mu = a \mu_1 + (1-a) \mu_2 &\ge& \mu_1
\end{aligned}

#### Show that $\sigma^2 \le \sigma_1^2$ and $\sigma^2 \le \sigma_2^2$:

\begin{aligned}
\sigma^2 &= \frac{\sigma_1^2 \sigma_2^2}{\sigma_1^2 + \sigma_2^2} \\
&= \frac{\frac{1}{\sigma_2^2}}{\frac{1}{\sigma_2^2}} \frac{\sigma_1^2 \sigma_2^2}{\sigma_1^2 + \sigma_2^2} \\
&= \frac{\sigma_1^2}{1 + \frac{\sigma_1^2}{\sigma_2^2}} \\
&\le& \sigma_1^2
\end{aligned}

because $\frac{\sigma_1^2}{\sigma_2^2} \ge 0$.  Replace $\sigma_2^2$ with $\sigma_1^2$ in the first step above, and it easily follows that $\sigma^2 \le \sigma_2^2$.

### Problem 9

This problem uses a number of "tricks" (identities that come from far afield).  I present these first:

#### Trick 1

\begin{aligned}
\mathbf{a}^T A^{-1} \mathbf{a} &= \mathbf{a}^T (A + B)^{-1} (A + B) A^{-1} \mathbf{a} \\
&= \mathbf{a}^T (A + B)^{-1} (I + B A^{-1}) \mathbf{a} \\
\mathbf{b}^T B^{-1} \mathbf{b} &= \mathbf{b}^T (A + B)^{-1} (A + B) B^{-1} \mathbf{b} \\
&= \mathbf{b}^T (A + B)^{-1} (I + AB^{-1}) \mathbf{b}
\end{aligned}

#### Trick 2

\begin{aligned}
A^{-1}(A^{-1} + B^{-1})^{-1} B^{-1} = (A+B)^{-1}
\end{aligned}

This is true because:
\begin{aligned}
A^{-1}(A^{-1} + B^{-1})^{-1} B^{-1}(A+B) &= (A+B)^{-1} (A+B) \\
A^{-1}(A^{-1} + B^{-1})^{-1} (B^{-1} A+I) &= I \\
A^{-1}(A^{-1} + B^{-1})^{-1} B^{-1} A &= I - A^{-1}(A^{-1} + B^{-1})^{-1} \\
(A^{-1} + B^{-1})^{-1} B^{-1} A &= A - (A^{-1} + B^{-1})^{-1} \\
(A^{-1} + B^{-1}) (A^{-1} + B^{-1})^{-1} B^{-1} A &= (A^{-1} + B^{-1}) A - (A^{-1} + B^{-1})(A^{-1} + B^{-1})^{-1} \\
B^{-1} A &= (A^{-1} + B^{-1}) A - I \\
&= I + B^{-1} A - I \\
&= B^{-1} A
\end{aligned}

_Note_:  Because $A, B$ are covariance matrices they are symmetric positive-definite.  Therefore,

\begin{aligned}
B^{-1}(A^{-1} + B^{-1})^{-1} A^{-1} &= \bigl ( B^{-1}(A^{-1} + B^{-1})^{-1} A^{-1} \bigr )^T \\
&= A^{-T}(A^{-1} + B^{-1})^{-T} B^{-T} \\
&= A^{-1}(A^{-1} + B^{-1})^{-1} B^{-1}
\end{aligned}

#### Trick 3

\begin{aligned}
A^{-1}(A^{-1} + B^{-1})^{-1} A^{-1} = (A+B)^{-1}BA^{-1}
\end{aligned}

Use Trick 2:
\begin{aligned}
A^{-1}(A^{-1} + B^{-1})^{-1} A^{-1} &= A^{-1}(A^{-1} + B^{-1})^{-1} B^{-1} B A^{-1} \\
&= (A+B)^{-1} BA^{-1}
\end{aligned}

#### Trick 4

\begin{aligned}
B^{-1}(A^{-1} + B^{-1})^{-1} B^{-1} = (A+B)^{-1}AB^{-1}
\end{aligned}

This is true because:
\begin{aligned}
B^{-1}(A^{-1} + B^{-1})^{-1} B^{-1} &= (A+B)^{-1}AB^{-1} \\
B^{-1}(A^{-1} + B^{-1})^{-1} &= (A+B)^{-1}A \\
B^{-1}(A^{-1} + B^{-1})^{-1}(A^{-1} + B^{-1}) &= (A+B)^{-1}A (A^{-1} + B^{-1}) \\
B^{-1} &= (A+B)^{-1}(I + AB^{-1}) \\
(A+B)B^{-1} &= (A+B)(A+B)^{-1}(I + AB^{-1}) \\
AB^{-1} + I &= I + AB^{-1}
\end{aligned}

### The solution

Start with the result of Problem 7:

\begin{aligned}
\text{(normalize the final density)} \ \kappa &= \kappa_3 \cdot (2 \pi)^{D/2} |(A^{-1} + B^{-1})^{-1}|^{1/2} \\
\kappa_3 &= \kappa_2 \cdot \exp[0.5 (A^{-1} \mathbf{a} + B^{-1} \mathbf{b})^T(A^{-1} + B^{-1})^{-1} (A^{-1} \mathbf{a} + B^{-1} \mathbf{b}) ] \\
\kappa_2 &= \kappa_1 \cdot \exp[-0.5 ( \mathbf{a}^T A^{-1} \mathbf{a} + \mathbf{b}^T B^{-1} \mathbf{b}) ] \\
\text{(normalize the initial density)} \ \kappa_1 &= \frac{1}{(2 \pi)^D |A|^{1/2} |B|^{1/2}}
\end{aligned}

$\kappa$ has the form of $\kappa = C \exp[ -0.5 \cdot D ]$.  Because there are many different ways to manipulate the determinant, start with the term inside the exponential to find the correct covariance form:

\begin{aligned}
D &= \mathbf{a}^T A^{-1} \mathbf{a} + \mathbf{b}^T B^{-1} \mathbf{b} - (A^{-1} \mathbf{a} + B^{-1} \mathbf{b})^T(A^{-1} + B^{-1})^{-1} (A^{-1} \mathbf{a} + B^{-1} \mathbf{b}) \\
\text{(expand)} &= \mathbf{a}^T A^{-1} \mathbf{a} + \mathbf{b}^T B^{-1} \mathbf{b} - \mathbf{a}^T A^{-1} (A^{-1} + B^{-1})^{-1} A^{-1} \mathbf{a} \\ 
&- \mathbf{a}^T A^{-1} (A^{-1} + B^{-1})^{-1} B^{-1} \mathbf{b} - \mathbf{b}^T B^{-1} (A^{-1} + B^{-1})^{-1} A^{-1} \mathbf{a} - \mathbf{b}^T B^{-1} (A^{-1} + B^{-1})^{-1} B^{-1} \mathbf{b} \\
\text{(apply Trick 1)} &= \mathbf{a}^T (A + B)^{-1} (I + B A^{-1}) \mathbf{a} + \mathbf{b}^T (A + B)^{-1} (I + AB^{-1}) \mathbf{b} - \mathbf{a}^T A^{-1} (A^{-1} + B^{-1})^{-1} A^{-1} \mathbf{a} \\ 
&- \mathbf{a}^T A^{-1} (A^{-1} + B^{-1})^{-1} B^{-1} \mathbf{b} - \mathbf{b}^T B^{-1} (A^{-1} + B^{-1})^{-1} A^{-1} \mathbf{a} - \mathbf{b}^T B^{-1} (A^{-1} + B^{-1})^{-1} B^{-1} \mathbf{b} \\
&= \mathbf{a}^T (A + B)^{-1} \mathbf{a} + \mathbf{a}^T (A + B)^{-1} B A^{-1} \mathbf{a} + \mathbf{b}^T (A + B)^{-1} \mathbf{b} + \mathbf{b}^T (A + B)^{-1} AB^{-1} \mathbf{b} - \mathbf{a}^T A^{-1} (A^{-1} + B^{-1})^{-1} A^{-1} \mathbf{a} \\ 
&- \mathbf{a}^T A^{-1} (A^{-1} + B^{-1})^{-1} B^{-1} \mathbf{b} - \mathbf{b}^T B^{-1} (A^{-1} + B^{-1})^{-1} A^{-1} \mathbf{a} - \mathbf{b}^T B^{-1} (A^{-1} + B^{-1})^{-1} B^{-1} \mathbf{b} \\
\text{(apply other tricks)} &= \mathbf{a}^T (A + B)^{-1} \mathbf{a} + \mathbf{a}^T (A + B)^{-1} B A^{-1} \mathbf{a} + \mathbf{b}^T (A + B)^{-1} \mathbf{b} + \mathbf{b}^T (A + B)^{-1} AB^{-1} \mathbf{b} \\
